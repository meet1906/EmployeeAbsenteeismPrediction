# Create Target

```python
data_preprocessed['Absenteeism Time in Hours'].median() #more than 3 - 1 less than 3 - 0
```

```python
targets = np.where(data_preprocessed['Absenteeism Time in Hours']>3,1,0)
```

```python
data_with_targets = data_preprocessed.drop(['Absenteeism Time in Hours'],axis=1)
data_with_targets
```

# Standardising

```python
from sklearn.preprocessing import StandardScaler

abstanteeism_scaler = StandardScaler()
```

```python
abstanteeism_scaler.fit(unscaled_inputs)
```

```python
scaled_inputs = abstanteeism_scaler.transform(unscaled_inputs)
```

# Splitting data into train test and then shuffle

```python
from sklearn.model_selection import train_test_split
```

```python
train_test_split(scaled_inputs,targets)
```

```python
x_train, x_test, y_train, y_test = train_test_split(scaled_inputs,targets, train_size = 0.8, random_state=20)
```

# Logistic Regression 

```python
from sklearn.linear_model import LogisticRegression
from sklearn import metrics
```

```python
reg = LogisticRegression()
reg.fit(x_train,y_train)
```

```python
reg.score(x_train,y_train)
```

# Finding intercepts and coeff

```python
reg.intercept_
```

```python
reg.coef_
```

```python
unscaled_inputs.columns.values
```

```python
feature_name = unscaled_inputs.columns.values
```

```python
feature_name
```

```python
summary_table = pd.DataFrame(columns=['Feature name'],data=feature_name)
summary_table['coefficient']=np.transpose(reg.coef_)
```

```python
summary_table
```

```python
summary_table.index = summary_table.index + 1
```

```python
summary_table.loc[0] = ['Intercept',reg.intercept_[0]]
summary_table = summary_table.sort_index()
summary_table
```

# Interpreting Coefficients

```python
#these are all log (odds) now we find their exp
summary_table['Odds Ratio']=np.exp(summary_table.coefficient)
```

```python
summary_table
```

```python
summary_table.sort_values('Odds Ratio',ascending=False)
```

# TEST MODEL

```python
reg.score(x_test,y_test)
```
